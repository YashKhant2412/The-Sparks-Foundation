## Task4 : Prediction using Decision Tree

```python
#---Importing Libraries---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.tree import plot_tree
from matplotlib.colors import ListedColormap
```

```python
#---Load Data and visulize---
df = pd.read_csv("Iris.csv")
df.head()
```

```python
#---Saperate Features and Labels---
X = df.iloc[:,1:5].values
y = df.iloc[:,-1].values
```

```python
#---Labelize categorize data
label = LabelEncoder()
y = label.fit_transform(y)
```

```python
#---Spliting Data into test and train---
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=24)
```

```python
#---Fitting data into Decision Tree
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
#---Prediction on test data
y_pred = classifier.predict(X_test)
```

```python
#---Find Accuracy score of prediction---
acc = metrics.accuracy_score(y_pred,y_test)
print(acc)
```

```python
#---Visulize Decision Tree
fig = plt.figure(figsize=(25,20))
_ = plot_tree(classifier,feature_names=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"],class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],filled =True)
```

```python
#---Visulize train data on Scatter Plot
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,2],
                color = ListedColormap(('blue','green','red'))(i), label = j)
plt.title("Decision Tree Classifier (TRAIN SET)")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```

```python
X_set, y_set = X_test, y_test

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,2],
                color = ListedColormap(('blue','green','red'))(i), label = j)
plt.title("Decision Tree Classifier (TEST SET)")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.legend()
plt.show()
```
